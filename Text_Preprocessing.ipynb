```python
import requests
import pandas as pd
import time
```

```python
API_KEY = "8265bd1679663a7ea12ac168da84d2e8"

genre_url = f"https://api.themoviedb.org/3/genre/movie/list?api_key={API_KEY}&language=en-US"
genre_response = requests.get(genre_url).json()

genre_dict = {g["id"]: g["name"] for g in genre_response["genres"]}
```

```python
all_movies = []

for page in range(1, 472):
    url = f"https://api.themoviedb.org/3/movie/top_rated?api_key={API_KEY}&language=en-US&page={page}"
    response = requests.get(url).json()

    for movie in response["results"]:
        overview = movie["overview"]
        genre_ids = movie["genre_ids"]

        if overview and genre_ids:
            genre_name = genre_dict.get(genre_ids[0])  # pick first genre

            all_movies.append({
                "description": overview,
                "genre": genre_name
            })

    time.sleep(0.2)  # avoid API rate limit
```

```python
import pandas as pd
import re
import nltk
```

```python
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
df = pd.read_csv("tmdb_multiclass_dataset.csv")

df.head()
```

```python
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()
```

```python
def preprocess_text(text):
    # 1. Lowercasing
    text = text.lower()

    # 2. Remove punctuation & special characters
    text = re.sub(r'[^a-z\s]', '', text)

    # 3. Tokenization
    tokens = text.split()

    # 4. Remove stopwords
    tokens = [word for word in tokens if word not in stop_words]

    # 5. Lemmatization
    tokens = [lemmatizer.lemmatize(word) for word in tokens]

    return " ".join(tokens)
```

```python
df["clean_description"] = df["description"].apply(preprocess_text)

df.head()
```
