```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
```

# Loading Dataset

```python
iris = load_iris()
X = iris.data.T #size is (4,150)
Y = np.array([1 if target ==0  else -1 for target in iris.target])
```

# Adding column of ones

```python
X_aug = np.vstack([X, np.ones(X.shape[1])]) #now size is (5,150)
```

#Solve least squares : minimising || y-X^T*beta||^2

```python
beta, _, _, _=np.linalg.lstsq(X_aug.T, Y, rcond=None)
```

#Predict

```python
y_pred = np.sign(X_aug.T @ beta)
```

#Confusion Matrix

```python
cm = confusion_matrix(Y, y_pred, labels = [1,-1])
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=["setosa (+1)", "Others (-1)"])
disp.plot(cmap=plt.cm.Blues)
plt.title("Least Squares Classifier (Confusion Matrix)")
plt.show()

print("Confusion Matrix : \n", cm)
```

Report:
The least squares classifier here, separates iris-setosa from other two species (iris-versicolor and iris-virginica). The confusion matrix shows that all setosa samples are correctly classified as +1, and other species are marked as -1.
